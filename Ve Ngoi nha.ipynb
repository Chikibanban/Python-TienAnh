```python
# Setting things up
import turtle
house=turtle.Turtle()
house.shapesize(1)
house.shape("turtle")
house.pensize(2)
house.speed(0)

# Drawning the triangle

for i in range (3):
    house.lt(120)
    house.fd(150)

#Drawning the paralellogram

house.fd(300)
house.lt(120)
house.fd(150)
house.lt(60)
house.fd(300)

#Drawning the rectangle
house.lt(120)
house.fd(150)
house.rt(30)

a = 300
b = 90
for i in range (4):
    house.fd(a)
    house.lt(b)

house.rt(b)
house.fd(150)
house.lt(b)
house.fd(235)
house.lt(66.25)
house.fd(165)

#3D the thing
house.bk(165)
house.lt(23)

for i in range (10):
    house.fd(10)
    house.penup()
    house.fd(10)
    house.pendown()
    house.fd(10)

house.lt(90)
for i in range (8):
    house.fd(10)
    house.penup()
    house.fd(10)
    house.pendown()
    house.fd(10)

house.rt(28)

for i in range (5):
    house.fd(10)
    house.penup()
    house.fd(10)
    house.pendown()
    house.fd(10)

house.penup()
house.goto(150,-240)
house.pendown()
house.rt(83)

for i in range (5):
    house.fd(10)
    house.penup()
    house.fd(10)
    house.pendown()
    house.fd(10)

#drawning the windows
house.penup()
house.goto(80,-150)
house.pendown()
house.lt(110)

for i in range (4):
    house.fd(60)
    house.lt(90)

house.fd(30)
house.lt(90)
house.fd(60)
for i in range (2):
    house.rt(90)
    house.fd(30)
house.rt(90)
house.fd(60)

house.penup()
house.goto(250,-90)    
house.pendown()

for i in range (4):
    house.fd(60)
    house.rt(90)

house.fd(30)
house.rt(90)
house.fd(60)

for i in range (2):
    house.lt(90)
    house.fd(30)

house.lt(90)
house.fd(60)

house.penup()
house.goto(120,-200)
house.pendown()

for i in range (2):
    house.rt(90)
    house.fd(60)
    house.rt(90)
    house.fd(100)


#Ending the code
turtle.exitonclick()


```

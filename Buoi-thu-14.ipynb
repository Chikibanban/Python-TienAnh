```python
def Fibonaci_Cach1(n):
    if n < 0:
        print('Loi nhap lieu')
    elif n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        return Fibonaci_Cach1(n-1) + Fibonaci_Cach1(n-2)
def Fibonaci_Cach2(n):
    a = 0
    b = 1
    if n < 0: print('Loi nhap lieu')
    elif n == 0: return 0
    elif n == 1 or n == 2: return b
    else: 
        for i in range(1,n):
            c = a + b
            a = b
            b = c 
        return c
if __name__ == "__main__":
    Fibonaci_Cach1(n)
    Fibonaci_Cach2()
```
